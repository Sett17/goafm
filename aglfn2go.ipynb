```python
code_first = '''package goafm
// Code generated by aglfn2go.ipynb. DO NOT EDIT.

'''
code_before = '''

// NamePair is a pair of glyph name and character name.
type NamePair struct {
	short string
	long  string
}

// contains returns true if the pair contains the given string.
func (np NamePair) contains(s string) bool {
	return np.short == s || np.long == s
}

// AGLFN is the Adobe Glyph List for New Fonts.
var AGLFN = map[uint16]NamePair{
'''
code_after = '''}
'''
```

```python
link = 'https://raw.githubusercontent.com/adobe-type-tools/agl-aglfn/master/aglfn.txt'
output = 'aglfn.go'
```

```python
comments = ""
import urllib.request

for line in urllib.request.urlopen(link):
    s = line.decode('utf-8')
    if s.startswith('#'):
        comments += "//" + s[1:]
    else:
        break
```

```python
import pandas as pd

df = pd.read_csv(link, sep=';', header=None, comment='#')
df.columns = ['code', 'short', 'long']
df
```

```python
with open(output, 'w') as f:
    f.write(code_first)
    f.write(comments)
    f.write(code_before)
    for i, row in df.iterrows():
        # print(f'\t0x{row["code"]}: {{"{row["short"]}", "{row["long"]}"}},')
        f.write(f'\t0x{row["code"]}: {{"{row["short"]}", "{row["long"]}"}},\n')
    f.write(code_after)
```
